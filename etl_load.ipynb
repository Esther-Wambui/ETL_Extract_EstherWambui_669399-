```python
import pandas as pd#for data manipulation
from pathlib import Path#for file path handling

Path("loaded_data").mkdir(exist_ok=True)#created the folder
```

## loading full transtransformed data

```python
full_df = pd.read_csv("transformed_full.csv")#reading the full transformed data
full_df.to_parquet("loaded_data/full_data.parquet", index=False)#saving the full transformed data to parquet format
pd.read_parquet("loaded_data/full_data.parquet").head()#previewing the full transformed data
```
